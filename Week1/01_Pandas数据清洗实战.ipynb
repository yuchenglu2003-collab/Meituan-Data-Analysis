# Pandas 数据清洗实战

**学习目标：** 掌握 Pandas 的增删改查、去重、数据类型转换，对大众点评原始评论数据进行初步清洗。

**产出：** 清洗后的干净数据集 `clean_data.csv`

## 1. 环境准备与数据加载

```python
import pandas as pd
import numpy as np
import os

# 设置显示选项
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

DATA_PATH = 'data1.csv'
df = pd.read_csv(DATA_PATH)
print(f'原始数据形状: {df.shape}')
df.head()
```

使用 `info()`、`describe()`、`isnull()` 了解数据结构与质量。

```python
# 查看各字段唯一值数量
df.nunique()
```

## 3. 增删改查实战

### 3.1 查 — 条件筛选与列选择

```python
# 筛选高评分评论（stars >= 4.5）
high_rating = df[df['stars'] >= 4.5]
print(f'高评分评论数: {len(high_rating)}')

# 选择特定列
df[['cus_id', 'shopID', 'stars', 'comment_time']].head()
```

```python
# 按门店分组统计评论数
shop_stats = df.groupby('shopID').agg(
    评论数=('cus_id', 'count'),
    平均评分=('stars', 'mean'),
    平均评论长度=('comment_len', 'mean')
).round(2)
shop_stats.sort_values('评论数', ascending=False)
```

### 3.2 — 新增衍生列

```python
# 新增评分等级列
def rating_level(star):
    if pd.isna(star):
        return '未知'
    if star >= 4.5:
        return '好评'
    elif star >= 3.5:
        return '中评'
    else:
        return '差评'

df['rating_level'] = df['stars'].apply(rating_level)
df['rating_level'].value_counts()
```

```python
# 新增评论时段列
def time_period(hour):
    if 6 <= hour < 11:
        return '上午'
    elif 11 <= hour < 14:
        return '中午'
    elif 14 <= hour < 18:
        return '下午'
    elif 18 <= hour < 22:
        return '晚上'
    else:
        return '深夜/凌晨'

df['time_period'] = df['hour'].apply(time_period)
df['time_period'].value_counts()
```

### 3.3 改 — 修改数据值

```python
# 将 comment_star 中的 NAN 字符串统一处理
print('修改前:', df['comment_star'].value_counts().get('NAN', 0))
df.loc[df['comment_star'] == 'NAN', 'comment_star'] = np.nan
print('修改后 NAN 数量:', df['comment_star'].isna().sum())
```

### 3.4 删 — 删除列或行

```python
# 演示：删除临时列（清洗流程中会保留有用列）
demo_df = df.copy()
demo_df = demo_df.drop(columns=['time_period'])
print(f'删除列后形状: {demo_df.shape}')
```

## 4. 去重

```python
print(f'重复行数量: {df.duplicated().sum()}')
print(f'完全重复行: {df.duplicated(keep=False).sum()}')

# 基于用户+门店+评论时间去重（业务逻辑去重）
dup_mask = df.duplicated(subset=['cus_id', 'shopID', 'comment_time'], keep='first')
print(f'业务重复评论: {dup_mask.sum()}')
```

## 5. 数据类型转换

```python
# comment_time 转为 datetime
df['comment_time'] = pd.to_datetime(df['comment_time'])
print(df['comment_time'].dtype)
print(f'时间范围: {df["comment_time"].min()} ~ {df["comment_time"].max()}')
```

```python
# comment_star 映射为数值评分
star_map = {
    'sml-str50': 5.0,
    'sml-str40': 4.0,
    'sml-str30': 3.0,
    'sml-str20': 2.0,
    'sml-str10': 1.0
}
df['star_from_comment'] = df['comment_star'].map(star_map)

# 用映射值填充 stars 缺失
df['stars'] = df['stars'].fillna(df['star_from_comment'])
print(f'stars 缺失数（填充后）: {df["stars"].isna().sum()}')
```

```python
# shopID 转为字符串类型（便于后续合并）
df['shopID'] = df['shopID'].astype(str)
df.dtypes
```

## 6. 处理缺失值

```python
# cus_comment 缺失：用空字符串填充
df['cus_comment'] = df['cus_comment'].fillna('')

# 维度评分缺失检查
for col in ['kouwei', 'huanjing', 'fuwu']:
    print(f'{col} 缺失: {df[col].isna().sum()}')

print(f'\n清洗后总缺失: {df.isnull().sum().sum()}')
```

## 7. 处理异常值

```python
# 评论长度为 0 的异常记录
zero_len = df[df['comment_len'] == 0]
print(f'评论长度为0的记录: {len(zero_len)}')
zero_len[['cus_id', 'cus_comment', 'comment_len', 'stars']]
```

```python
# 使用 IQR 方法检测评论长度异常值
Q1 = df['comment_len'].quantile(0.25)
Q3 = df['comment_len'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['comment_len'] < lower_bound) | (df['comment_len'] > upper_bound)]
print(f'评论长度异常值数量: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)')
print(f'正常范围: [{max(0, lower_bound):.0f}, {upper_bound:.0f}]')
```

```python
# 评分异常值检查（应在 1-5 之间）
invalid_stars = df[(df['stars'] < 1) | (df['stars'] > 5)]
print(f'评分超出范围记录: {len(invalid_stars)}')
```

## 8. 综合清洗流程

```python
def clean_dianping_data(raw_df):
    """大众点评评论数据清洗流水线"""
    data = raw_df.copy()
    
    # 1. comment_star 标准化
    data.loc[data['comment_star'] == 'NAN', 'comment_star'] = np.nan
    
    # 2. 时间类型转换
    data['comment_time'] = pd.to_datetime(data['comment_time'])
    
    # 3. 评分填充
    star_map = {'sml-str50': 5.0, 'sml-str40': 4.0, 'sml-str30': 3.0,
                'sml-str20': 2.0, 'sml-str10': 1.0}
    data['stars'] = data['stars'].fillna(data['comment_star'].map(star_map))
    
    # 4. 缺失值处理
    data['cus_comment'] = data['cus_comment'].fillna('')
    
    # 5. 去重
    data = data.drop_duplicates(subset=['cus_id', 'shopID', 'comment_time'], keep='first')
    data = data.drop_duplicates(keep='first')
    
    # 6. 异常值过滤：保留有效评分和合理评论长度
    data = data[(data['stars'] >= 1) & (data['stars'] <= 5)]
    data = data[data['comment_len'] >= 0]
    
    # 7. 衍生列
    data['rating_level'] = data['stars'].apply(
        lambda s: '好评' if s >= 4.5 else ('中评' if s >= 3.5 else '差评')
    )
    data['shopID'] = data['shopID'].astype(str)
    
    return data.reset_index(drop=True)

# 重新加载原始数据执行清洗
raw_df = pd.read_csv(DATA_PATH)
clean_df = clean_dianping_data(raw_df)
print(f'原始: {raw_df.shape} -> 清洗后: {clean_df.shape}')
clean_df.head()
```

## 9. 保存清洗后数据

```python
OUTPUT_PATH = 'clean_data.csv'
clean_df.to_csv(OUTPUT_PATH, index=False, encoding='utf-8-sig')
print(f'已保存至 {OUTPUT_PATH}')
print(f'文件大小: {os.path.getsize(OUTPUT_PATH)/1024/1024:.2f} MB')
```

## 10. 清洗前后对比总结

```python
summary = pd.DataFrame({
    '指标': ['总行数', '缺失stars', '重复行', '门店数', '平均评分', '平均评论长度'],
    '清洗前': [
        len(raw_df),
        raw_df['stars'].isna().sum(),
        raw_df.duplicated().sum(),
        raw_df['shopID'].nunique(),
        round(raw_df['stars'].mean(), 2),
        round(raw_df['comment_len'].mean(), 1)
    ],
    '清洗后': [
        len(clean_df),
        clean_df['stars'].isna().sum(),
        clean_df.duplicated().sum(),
        clean_df['shopID'].nunique(),
        round(clean_df['stars'].mean(), 2),
        round(clean_df['comment_len'].mean(), 1)
    ]
})
summary
```
